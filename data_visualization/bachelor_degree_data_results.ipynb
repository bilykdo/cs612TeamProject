```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
outputData = pd.read_csv('../data_analysis/bachelor_degree_resultOfInvestmentReturnBasedOnMajorByType.csv')

outputData
```

```python
sb.catplot(x="Average Percent Of Salary", y="Major Type", data=outputData)
```

```python
sb.lmplot(x='Average Percent Of Salary', y='Starting Median Salary', data=outputData)
```

```python
sb.lmplot(x='Average Percent Of Salary', y='Starting Median Salary', data=outputData,
           fit_reg=False, # No regression line
           hue='Major Type')   # Color by percentage
```

```python
# Find a median value by major type for Retirement Savings In 10 Years
df = outputData.pivot_table(index='Major Type', values='Retirement Savings In 10 Years', aggfunc=np.median).sort_values(by ='Retirement Savings In 10 Years', ascending=False)
sb.heatmap(df, annot=True, fmt=".1f")
```
